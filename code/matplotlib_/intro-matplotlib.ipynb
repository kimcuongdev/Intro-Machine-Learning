## Pyplot API vs Object Orient API

* Pyplot API -> Quickly
* OO API -> Advance

### Pyplot API

```python
x = np.linspace(0,10,1000) #1000 s t 0 - 10
x
```

```python
plt.plot(x, np.sin(x))
```

#### t tn cc trc

```python
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.plot(x,np.sin(x))
```

#### Gii hn cc trc

```python
plt.xlim([0,4])
plt.ylim([-0.75,0.75])
plt.plot(x,np.sin(x))
```

```python
xmin = 0
xmax = 4
ymin = -0.7
ymax = 0.6
plt.axis([xmin,xmax,ymin,ymax])
plt.plot(x, np.sin(x))
```

#### i Style ng

```python
plt.plot(x, np.sin(x), linestyle="dashed", color="blue")
```

#### V hai  th

```python
plt.ylabel("f(x)")
plt.plot(x, np.sin(x), color="blue",linestyle="dashed", label="sin(x)")
plt.plot(x, np.cos(x), color="red", label="cos(x)")
plt.legend()
plt.title("sin(x) and cos(x)")
```

### Object Orient API

```python
## 1. Prepare Data
x = [1,2,3,4]
y = [11,22,33,44]

## 2.Setup plot
fig, ax = plt.subplots(figsize=(5,5)) # figure size = (width, height)

## 3.Plot the data
ax.plot(x,y)
ax.set(title="A simple plot",xlabel="x",ylabel="y = 11x")
```

#### OO API

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

size = rng.rand(100) * 1000
color = rng.rand(100)

fig, ax = plt.subplots()
img = ax.scatter(x,y, s = size, c=color, cmap='viridis', alpha=0.3)

fig.colorbar(img)
```

### Bar Chart

#### Vertical Bar Chart

```python
##dictionary
sorf_drink_prices = {"Coke": 10,
                     "Pepsi": 12,
                     "Sprite": 8}
fig, ax = plt.subplots()
ax.bar(sorf_drink_prices.keys(), sorf_drink_prices.values())
ax.set(title="Soft Drink Price", ylabel="Price($)")
```

#### Horizontial Chart

```python
fig, ax = plt.subplots()
ax.barh(sorf_drink_prices.keys(), sorf_drink_prices.values())
```

### Histogram Chart

```python
# Prepare Data
np.random.seed(42)
student_height = np.random.normal(170, 10, 250) ##(mean, std, num)
```

#### Pyplot API

```python
plt.hist(student_height, bins=50)
```

## Subplots

```python
x = np.linspace(0,10,100)
```

### OO API (Option 1)

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
```

```python
## PLot Data
ax1.plot(x, x/2)
ax1.plot(x, x*2)
ax1.plot(x, x**2)

ax2.scatter(np.random.random(10), np.random.random(10))

ax3.bar(sorf_drink_prices.keys(), sorf_drink_prices.values())

ax4.hist(student_height)

fig
```

### OO API (Option 2)

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
## PLot Data
ax[0,0].plot(x, x/2)
ax[0,0].plot(x, x*2)
ax[0,0].plot(x, x**2)

ax[0,1].scatter(np.random.random(10), np.random.random(10))

ax[1,0].bar(sorf_drink_prices.keys(), sorf_drink_prices.values())

ax[1,1].hist(student_height)
```
